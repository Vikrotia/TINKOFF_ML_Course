#### Зависимость между потенциалом и Международным Рейтингом футболистов
По данным, представленным на графике ниже, можно проследить что есть прямая зависимость между этими характеристиками. Чем выше больше международный рейтинг футболистов, тем выше их потенциал.

```python
potential = sorted(df['Potential'].dropna().unique().tolist())
new_df2 = pd.DataFrame(potential).rename(columns = {0:'potential'})
for j in range(1, 6):
    count_potential = []
    for i in potential:
        try:
            count_potential.append(df.loc[df['International Reputation'] == j,'Potential'].value_counts()[i])
        except KeyError:
            count_potential.append(0)
    new_df2[j] = count_potential

new_df2['mid'] = new_df2[[1,2,3,4,5]].mean(axis = 1)
new_df2['mid1'] = new_df2['potential'] * new_df2[1]
sum1 = new_df2['mid1'].sum() / new_df2[1].sum()
new_df2['mid2'] = new_df2['potential'] * new_df2[2]
sum2 = new_df2['mid2'].sum() / new_df2[2].sum()
new_df2['mid3'] = new_df2['potential'] * new_df2[3]
sum3 = new_df2['mid3'].sum() / new_df2[3].sum()
new_df2['mid4'] = new_df2['potential'] * new_df2[4]
sum4 = new_df2['mid4'].sum() / new_df2[4].sum()
new_df2['mid5'] = new_df2['potential'] * new_df2[5]
sum5 = new_df2['mid5'].sum() / new_df2[5].sum()
potential_sum = [sum1, sum2, sum3, sum4, sum5]
potential_sum_count = [1,2,3,4,5]
fig = px.line(x=potential_sum, 
    y=potential_sum_count, 
    title='Зависимость между потенциалом и Международным Рейтингом футболистов')
fig.update_layout(title_font_size = 20, 
    xaxis_title = "Потенциал",
    yaxis_title = "Рейтинг")
fig.show()
```

12. Зависит ли **стоимость** игрока (Value) от его "**финтов**" (Skill Moves)?

По даннным графика ниже можно сказать, что прослеживается некоторая зависимость между стоимостью игрока и его финтами, т.к график движется вверх с возрастением стоимости. Но достаточно много выбросов, которые выделяются из общей тенденции. поэтому прямой зависимости между этими показателями нет.

26. Насколько средняя зарплата (Wage) игроков с рейтингом (Overall) 90+ больше средней зарплаты с рейтингом [80 - 90]? Визуализируйте возможную зависимость

```python
hh = df[['Overall', 'Wage']].dropna()
lst = hh['Wage'].tolist()
lst_num = []
for i in lst:
    c = i.replace('€', '')
    if('M' in i): c = float(c.replace('M', '')) * 1000000
    elif ('K' in i): c = float(c.replace('K', '')) * 1000
    lst_num.append(c)
hh['Wage_norm'] = lst_num
hh2 = hh.loc[((hh['Overall'] <= 90) & (hh['Overall'] >= 80))]
hh3 = hh.loc[((hh['Overall'] >= 90))]
hh2 = hh2.astype({'Wage_norm': np.int64})
hh2['Wage_norm'].mean()
hh3 = hh3.astype({'Wage_norm': np.int64})
hh3['Wage_norm'].mean()
value = [hh2['Wage_norm'].mean(), hh3['Wage_norm'].mean()]
name = ['Overall [80 - 90]', 'Overall 90+']
fig = px.bar(x=name, y=value, title='Зависимость между продолжительностью контракта игрока и его стоимостью')
fig.show()
```

```python
print(f"В {round(value[1]/value[0], 2)} раз средняя зарплата игроков с рейтингом 90+ больше средней зарплаты с рейтингом [80 - 90]")
```
